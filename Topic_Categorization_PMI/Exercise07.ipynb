```
import math
import re
from collections import defaultdict, Counter
from nltk import ngrams
import pandas as pd
import glob as gb
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
#this method does the tokenization and removal of numbers and punctuations
def tokenize(text):
    "List all the word tokens (consecutive letters) in a text. Normalize to lowercase."
    return re.findall('[a-z]+', text.lower())
```

```
#this method handles reading the training corpus
def readCorpus(directory="ag_news_csv/train.csv"):
  with open(directory, 'r',encoding="utf8", errors='replace') as doc:
      data = doc.readlines()
  return data
```

```
#read corpus method is invoked here
training_data=readCorpus()
```

```
#divide sentence to their categories 1,2,3,4 they are saaved respectively to list in index 0,1,2,3 which is a list of lists
catSentList=[[],[],[],[]]
for x in training_data:
  catSentList[int(x[1:2])-1].append(tokenize(x))
```

```
#merge category list gotten above to a universal list for each category
import more_itertools
reducedList=[]
for i in range(len(catSentList)):
  reducedList.append(list(more_itertools.flatten(catSentList[i])))
```

```
#count the words in each group/category
countDictList=[]
for i in range(len(reducedList)):
  countDictList.append(Counter(reducedList[i]))
```

```
#this takes the training data as a whole instead of filtering and grouping them 
#universal count
listofSentList=[]
for x in training_data:
  listofSentList.append(tokenize(x))
#convert the list of list into just one which is a universal list
flatSentList=list(more_itertools.flatten(listofSentList))
#count word occurence in the list and remove words that have counts less than 3
sentCount=Counter(flatSentList)
countGT3=Counter(dict(filter(lambda x: x[1] >= 3, sentCount.items())))
```

```
#get the count of all words in each category
sumofValuesPercat=[sum(dict(countDictList[0]).values()),sum(dict(countDictList[1]).values()),sum(dict(countDictList[2]).values()),sum(dict(countDictList[3]).values())]
#get the count of all words in the universal list
sumoGT3 = sum(dict(countGT3).values())
```

```
def AVGPMI(f):
  avgpmi = 0
  for i in range(len(catSentList)): #loop through each corpora category
    numerator = countDictList[i][f]/sumofValuesPercat[i]
    denomenator = countGT3[f]/sumoGT3
    if numerator==0:
      avgpmi+=0
    else:
      avgpmi = avgpmi + (0.25) * math.log(numerator/denomenator,2) # where the 0.25 in the equation the probability of that category
  return avgpmi 
```

```
#compute the PMI for the words with count >=3 using first method
pmicountGT3 = dict(countGT3) 
for key in pmicountGT3:
  pmicountGT3[key] = AVGPMI(key) 
```

```
#take the top 20 words based on the average pmi
pmicountGT3=Counter(pmicountGT3)
pmicountGT3.most_common(20)
```

```
def maxPMI(f):
  pmis=[]
  for i in range(len(catSentList)):
    numerator = countDictList[i][f]/sumofValuesPercat[i]
    denomenator = countGT3[f]/sumoGT3
    if numerator==0:
      pmis.append(0)
    else:
      pmis.append(math.log(numerator/denomenator,2))
      
  return max(pmis) 
```

```
#compute the PMI for the words with count >3 using the second method
pmicountGT3max = dict(countGT3)
for key in pmicountGT3max:
  pmicountGT3max[key] = maxPMI(key) 
```

```
#get the top 20 pmi and the words using the second method
pmicountGT3max=Counter(pmicountGT3max)
EPmi20=pmicountGT3max.most_common(20)
```
